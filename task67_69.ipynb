```python
import seaborn as sns
import pandas as pd
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
penguins = penguins.dropna()
```

```python
penguins.loc[penguins['bill_length_mm'] > 42, 'new_col'] = 'high'
penguins.loc[(penguins['bill_length_mm'] <= 42) & (penguins['bill_length_mm'] >= 35), 'new_col'] = 'middle'
penguins.loc[penguins['bill_length_mm'] < 35, 'new_col'] = 'low'
```

```python
penguins.head(15)
```

```python
sns.histplot(data = penguins, x = 'flipper_length_mm', hue = 'new_col')
```
