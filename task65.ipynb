```python
import seaborn as sns
import pandas as pd
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
penguins.shape
```

```python
penguins.isnull().sum()
```

```python
penguins = penguins.dropna()
```

```python
sns.scatterplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", size = "island")
```

```python
sns.scatterplot(data = penguins, x = "bill_length_mm", y = "bill_depth_mm", hue = "body_mass_g")
```

```python
sns.scatterplot(data = penguins, x = "body_mass_g", y = "bill_length_mm", hue="sex")
```

```python
t = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
g = sns.PairGrid(penguins[t])
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.scatterplot)
```

```python
sns.heatmap(penguins.corr())
```

```python
penguins.corr()
```

```python
sns.histplot(data = penguins, x = "species", hue = "sex")
```

```python
sns.histplot(data = penguins, x = "body_mass_g", hue = "species")
```
